```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import random
```

```python
graph = nx.generators.random_graphs.gnm_random_graph(n=100, m=200)
plt.figure(figsize=(15, 15))
plt.title('Graph Visualization')
nx.draw(graph, with_labels=True, node_color='yellow')
```

```python
adjacency_matrix = nx.adjacency_matrix(graph).todense()
adjacency_matrix
```

```python
adjacency_list = nx.to_dict_of_lists(graph)
adjacency_list
```

```python
def depth_first(graph, source):
    return nx.dfs_preorder_nodes(
        G=graph,
        source=source
    )

def breadth_first(graph, source, target):
    return nx.bidirectional_shortest_path(
        G=graph,
        source=source,
        target=target
    )
```

```python
depth_first_args = list(depth_first(graph, 0))

plt.figure(figsize=(10, 10))
plt.title('Graph')
nx.draw_shell(graph.subgraph(depth_first_args), with_labels=True, node_color='yellow')

print(depth_first_args)
```

```python
initial_node = random.randrange(len(graph)-1)
final_node = random.randrange(len(graph)-1)
breadth_first_args = list(breadth_first(graph, initial_node, final_node))

plt.figure(figsize=(10, 10))
plt.title('Shortest path')
nx.draw_shell(graph.subgraph(breadth_first_args), with_labels=True, node_color='yellow')

print(breadth_first_args)
```
